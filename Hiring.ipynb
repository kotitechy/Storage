```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/codebasics/py/master/ML/2_linear_reg_multivariate/Exercise/hiring.csv"
df = pd.read_csv(url)
df.head(9)
```

# Data Pre-processing

```python
df['experience'] = df['experience'].fillna('zero')
df
```

```python
# mean of experience
from word2number import w2n
df['experience']=df.experience.apply(w2n.word_to_num)
```

```python
df
```

```python
from math import floor
mv=floor(df['test_score(out of 10)'].mean())
df['test_score(out of 10)']=df['test_score(out of 10)'].fillna(mv)
df
```

#fitting


```python
from sklearn.linear_model import LinearRegression as lm
reg = lm()
```

```python
reg.fit(df[['experience','test_score(out of 10)','interview_score(out of 10)']],df['salary($)'])
```

```python
floor(reg.predict([[0,5,5]]))
```
